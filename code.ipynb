```python
# part b solution: comparing root means square error for both the years 2000 and 2001

import math
import numpy as np
from numpy.linalg import inv




# loading the index list as a dictionary with the line numbers in the text file 'nasdaq**' as keys

# for 2000

indexlist00=[]
with open('nasdaq00.txt') as f:
    for line in f:
        indexlist00.append(float(line.strip()))


#for 2001

indexlist01=[]
with open('nasdaq01.txt') as f:
    for line in f:
        indexlist01.append(float(line.strip()))


def coeffcalc(indexlist):    #takes the dictionary of indices on a particular year and gives the rms error
    T=len(indexlist)  # no. of samlples in the nasdaq** file
    K=np.zeros((3,3),dtype=float)
    A=np.zeros((3,1),dtype=float)
    B=np.zeros((3,1),dtype=float)
    
    # creating the vector B
    for i in range(1,4):
        s=0
        for t in range(3,T):   # we are summing from fourth day
            s=s+indexlist[t]*indexlist[t-i]
        B[i-1]=s
    

# creating the matrix K
    for i in range(3):
        for j in range(3):
            s=0
            for t in range(3,T):  # we are summing from fourth day
                    s=s+indexlist[t-j-1]*indexlist[t-i-1]
            K[i,j]=s   # K is a symmetric matrix by definition

    
            
# finding the optimal coefficients for maximizing the log-likelihood:
    Kinv=np.linalg.inv(K)
    A=np.matmul(Kinv,B)
    return A
    
def errcalc(A, indexlist):
    T=len(indexlist)  # no. of samlples in the nasdaq** file
    # calculating root mean square error
    err=0
    for i in range(3,T):
        err=err+np.power((indexlist[i]-A[0]*indexlist[i-1]-A[1]*indexlist[i-2]-A[2]*indexlist[i-3]),2)
    
    rmerr=np.sqrt(err/(T-3))
    return rmerr
    

A=coeffcalc(indexlist00)
print ('using the optimal coefficients {a_i}s from the data of 2000 \n')
error00=errcalc(A,indexlist00)
print(f"the root mean square error from the NASDAQ data of the year 2000 is {error00}")
error01=errcalc(A,indexlist01)
print(f"the root mean square error from the NASDAQ data of the year 2001 is {error01}")    
```
